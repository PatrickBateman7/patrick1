```python
#Lista 3 - Pedro Aubert Cordeiro -- 2110083
#link para o github https://github.com/PatrickBateman7/Lista-3-Pandas-PedroAubert
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
titanic = pd.read_csv("/content/drive/MyDrive/class_titanic.csv")
titanic = pd.DataFrame(titanic)
display(titanic)
```

```python
#Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)

titanic.info()
```

```python
#Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo

titanic['Fare'] = titanic['Fare'].replace(0,np.nan)

titanic['Fare'] = titanic['Fare'].dropna()
```

```python
#Calcule a média desta mesma coluna

média = titanic['Fare'].mean()
```

```python
#Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.
 
print(média)

ajuste1 = titanic['Fare'] >= 8.05
ajuste2 = titanic['Fare'] <= 146.5208

titanic = titanic[ajuste1]
titanic = titanic[ajuste2]
```

```python
#Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.

titanic2 = titanic.drop(['Age','Fare','Survived','Parch','Name'],axis = 1)
```

```python
#Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

titanic2['nova_coluna'] = titanic2['PassengerId']-titanic2['Pclass']

titanic2['nova_coluna'] = titanic2['nova_coluna']*0.8
```

```python
#Embaralhe aleatoriamente o dataframe (índices e valores).

titanic_random = titanic2.sample(frac=1)
```

```python
#Ordene o Dataframe seguindo um critério definido por você (Explique o critério)

titanic2 = titanic2[['nova_coluna','Pclass','PassengerId','Sex','Cabin']] #-> Organizei as colunas na ordem: primeiramente a nova coluna criada, depois a classe mostrada em Pclass, seguida pelo Id do passageiro e o sexo do mesmo e por último a cabine

titanic2.sort_values(by = ['Pclass']).reset_index(drop=True) # -> Organizei a classe socio-econômica dos passageiros (Pclass) em ordem crescente e depois ajustei o index para concordar com a nova organização.
```

```python
#Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da coluna.

total = titanic2.groupby(['Pclass','Cabin']).count()
```

```python
#Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)

titanic2['Sex'] = titanic2['Sex'].map({'male':'M','female':'F'})# -> Utilizei a função MAPS para alterar a coluna 'Sex', substituindo todas as strings 'male por 'M'(letra que representa o sexo masculino) e 'female' por 'F'(letra que representa o sexo feminino).
```
